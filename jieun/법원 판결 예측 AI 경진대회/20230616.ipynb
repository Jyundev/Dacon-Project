```python
import pandas as pd
import numpy as np

train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
display(train, test)
```

```python
print(train['facts'][0]) 
#! first_party랑 facts랑 단어 매치가 안됨
```

```python
all_data = pd.concat([train,test])
all_data
```

```python

#* TfidF 는 빈도수만 세서 그닥
#? 단어 빈도 보다 문맥 파악하는게 중요해 보임, 단어 빈도수로만은 한계 있을 듯
#* 세 칼럼을 그냥 공백 으로만 더하기에는 좀 그렇다. first_party가 여러 단어로 결합되어 있으므로 하나의 단어로 인식되도록 전처리 해줄 것 - 띄어쓰기 없애고 공백은 _(언더바)로 채워주기
#? 첫번째 행 데이터를 확인해보면 판결문에서는 TOMSON이나  SECOND_PARTY 에서는 풀네임이니 나중에 전처리해보면 좋을 것 같다.
```

```python
all_data['first_party'] = all_data['first_party'].apply(lambda x: x.replace(' ', '_'))
all_data['second_party'] = all_data['second_party'].apply(lambda x: x.replace(' ','_'))
#all_data['first_party'] = all_data['first_party'].apply(lambda x: x.split()).apply(lambda x: '_'.join(x))
all_data
```

```python
all_data['text'] = all_data['first_party'] + ' ' + all_data['second_party'] + ' ' + all_data['facts']
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

tk = Tokenizer() 
tk.fit_on_texts(all_data['text']) #* 특정 단어를 숫자로 등록 
text = tk.texts_to_sequences(all_data['text']) #* 각각의 등록된 단어를 숫자로 바꿔줌
pad_text = pad_sequences(text, maxlen = 200) #* 차원 통일
pad_text.shape
#* 단어가 996개 이면 너무 많음, 0이 불필요하게 너무 많이 들어감 -> 200으로 줄여봄 (하나의 데이터에 들어가는 단어 수를 200개로 제한)
```

```python
train2 = pad_text[:len(train)]
test2 =  pad_text[len(train):]
display(train2, test2)
```

```python
tk.word_index #* 이 종류의 개수가 단어의 종류의 개수
print(len(tk.word_index))
```

```python
from tensorflow.keras import *
from tensorflow.keras.models import *
from tensorflow.keras.layers import *
from tensorflow.keras.optimizers import *
from tensorflow.keras.callbacks import *
from sklearn.model_selection import train_test_split
from sklearn.model_selection import StratifiedKFold

skf = StratifiedKFold(n_splits=10, shuffle=True, random_state = 42)
result = 0
for train_index, valid_index in skf.split(train2, train['first_party_winner']):
    X_train, X_valid = train2[train_index], train2[valid_index]
    y_train, y_valid = train['first_party_winner'][train_index], train['first_party_winner'][valid_index]
    model = Sequential()
    model.add(Embedding(len(tk.word_index)+1, 10, input_length = 200)) #* (batch, 200, 10), 0이라는 단어가 추가 되어서 1을 추가해줌, 0은 자리만 차지함, 의미없는 단어임을 인식 시켜야함, Embedding 은 단어의 의미를 하나하나를 파악하기 위한 준비 단계, 10은 몇 가지의 의미로 구분해서 차원으로 받아 들일지 설정한 값, 임베딩은 2차원
    model.add(Flatten()) #* 2차원을 1차원으로 차원 축소
    model.add(Dense(2, activation ='softmax')) #* 클래스 2개 구분, 0과 1사이의 확률로 바꿈(softma)
    model.compile(loss = 'sparse_categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy']) #* sparse -> 숫자의 크기가 아닌 분류임을 명확히 알려줌
    es = EarlyStopping(patience = 5, restore_best_weights = True) #* 학습하다가 최적의 순간에서 멈춤
    model.fit(X_train, y_train, validation_data = (X_valid, y_valid), epochs = 100, batch_size = 32, callbacks = [es])
    result += model.predict(test2) / 10 
```
